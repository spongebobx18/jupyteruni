```python
#Replacing null values in each column.
```

```python
eachColumn = ['fixed acidity', 'volatile acidity', 'citric acid',
       'residual sugar', 'chlorides', 'free sulfur dioxide',
       'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol','quality']
```

```python
for column in eachColumn:
    data[column].fillna(data[column].mean(), inplace=True)
```

```python
#METHOD: Replacing null values with the mean of each column.

#Reason for replacing null values using the mean of each column:

#Replacing null values with the mean of the column is a common technique used in data analysis when missing data is present. Deleting rows with missing values can lead to loss of information and reduce the sample size, which may negatively impact the accuracy and representativeness of the analysis.

#By replacing null values with the mean of the column, we can preserve the sample size and provide a reasonable estimate of the missing data. This approach assumes that the data is normally distributed and that the missing values are missing at random, meaning that the probability of a value being missing is unrelated
```

```python
acids = ['fixed acidity', 'volatile acidity', 'citric acid']
```

```python
for acid in acids:
    plt.title(f'Correlation between the pH and {acid}')
    sn.scatterplot(x=acid ,y='pH' ,data=data)
    plt.figure(figsize=(12,10))
    plt.show()
```

```python
plt.title('A Graph To Show Frequency of Mode')
sn.histplot(data['fixed acidity'], bins=20)
plt.xlabel('Fixed Acidity')
plt.ylabel('Frequency')
plt.title('Distribution of Fixed Acidity')
plt.show()
```

```python
mean_fixed_acidity = data['fixed acidity'].mean()

compared_mean = data['fixed acidity'] < mean_fixed_acidity

plt.pie(compared_mean.value_counts(),labels =['Values less than mean', 'Values greater than mean'],colors = ['pink','black'] )
plt.title('Comparison between fixed acidity values and the mean')
plt.show()
plt.savefig('image.png')
```
